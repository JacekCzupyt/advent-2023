```python
def variations(s, l):
    if l == 1:
        return [np.array(s)]
    return [
        np.append(i, suf)
        for i in range(s+1)
        for suf in variations(s-i, l-1)
    ]
```

```python
poss = np.zeros(len(lines))

for k, (line, broken) in tqdm(enumerate(lines)):
    var = variations(len(line) - sum(broken) - len(broken) + 1, len(broken)+1)
    
    for v in var:
        success = True
        i = 0
        for j in range(len(broken)):
            if '#' in line[i:i+v[j]+(j!=0)]:
                success = False
                break
            i += v[j] + (j!=0)
            if '.' in line[i:i+broken[j]]:
                success = False
                break
            i += broken[j]
        if success and '#' not in line[i:]:
            poss[k] += 1
        
sum(poss) 
```

```python
def compute_possibilities(string, data):
    state_map = np.zeros((len(data)+1, len(string)-sum(data)+2-len(data)))
    state_map[0, 0] = 1
    for j in range(state_map.shape[0]):
        for i in range(state_map.shape[1]):
            pos = i + sum(data[:j]) + j - (j == state_map.shape[0]-1)
            if i < state_map.shape[1] - 1 and string[pos] != "#":
                state_map[j, i+1] += state_map[j, i]
            if j < state_map.shape[0] - 1 and "." not in string[pos:pos+data[j]] and (pos+data[j] >= len(string) or string[pos+data[j]] != "#"):
                state_map[j+1, i] += state_map[j, i]
    return state_map[-1, -1]

string, data = lines[4]
compute_possibilities(string, data)
```
