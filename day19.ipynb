```python
class Step:
    def __init__(self, desc):
        gr = re.match(r"(.)([=<>])([0-9]+)", desc).groups()
        self.prop = gr[0]
        self.op = gr[1]
        self.val = int(gr[2])
        
    def eval(self, item):
        val = item[self.prop]
        match self.op:
            case "=":
                return val == self.val
            case ">":
                return val > self.val
            case "<":
                return val < self.val
        
    def update_ranges(self, ranges):
        match self.op:
            case "=":
                new_arr = np.logical_and(np.arange(4001)==self.val, ranges[self.prop])
                if new_arr.sum():
                    yield (True, {**ranges, self.prop: new_arr})
                new_arr = np.logical_and(np.arange(4001)!=self.val, ranges[self.prop])
                if new_arr.sum():
                    yield (False, {**ranges, self.prop: new_arr})
            case ">":
                new_arr = np.logical_and(np.arange(4001) > self.val, ranges[self.prop])
                if new_arr.sum():
                    yield (True, {**ranges, self.prop: new_arr})
                new_arr = np.logical_and(np.arange(4001) <= self.val, ranges[self.prop])
                if new_arr.sum():
                    yield (False, {**ranges, self.prop: new_arr})
            case "<":
                new_arr = np.logical_and(np.arange(4001) < self.val, ranges[self.prop])
                if new_arr.sum():
                    yield (True, {**ranges, self.prop: new_arr})
                new_arr = np.logical_and(np.arange(4001) >= self.val, ranges[self.prop])
                if new_arr.sum():
                    yield (False, {**ranges, self.prop: new_arr})
```

```python
class WorkFlow:
    def __init__(self, desc):
        self.name, string = re.match(r"(.+)\{(.+)}", desc).groups()
        self.steps = []
        for step in string.split(','):
            parts = step.split(':')
            if len(parts) == 2:
                self.steps.append((Step(parts[0]), parts[1]))
            else:
                self.steps.append((None, parts[0]))
    
    def eval(self, item):
        for step in self.steps:
            if step[0] is None or step[0].eval(item):
                return step[1]
        return None
    
    def update_ranges(self, ranges):
        for step in self.steps:
            if step[0] is not None:
                for res in step[0].update_ranges(ranges):
                    if res[0]:
                        yield (step[1], res[1])
                    else:
                        ranges = res[1]
            else:
                yield (step[1], ranges)
```

```python
class Algorithm:
    def __init__(self, desc):
        workflows = [
            WorkFlow(line)
            for line in desc.split("\n")
        ]
        self.workflows = {
            w.name: w
            for w in workflows
        }
        
    def eval(self, item):
        wf = 'in'
        while wf not in ['A', 'R']:
            wf = self.workflows[wf].eval(item)
        return wf == 'A'
    
    def calculate_ranges(self, ranges):
        result = 0
        q = deque([('in', ranges)])
        while len(q) > 0:
            item = q.popleft()
            if item[0] == 'A':
                result += np.prod([val.sum() for val in item[1].values()])
            elif item[0] != 'R':
                for res in self.workflows[item[0]].update_ranges(item[1]):
                    q.append(res)
        return result
```
