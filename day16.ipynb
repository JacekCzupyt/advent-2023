```python
map = np.array([[c for c in line] for line in input.split("\n")])
```

```python
def nx(char, dir):
    match char:
        case '.':
            return [dir]
        case '/':
            return [-dir[::-1]]
        case '\\':
            return [dir[::-1]]
        case '-':
            if dir[0] == 0:
                return [dir]
            return [np.array([0, 1]), np.array([0, -1])]
        case '|':
            if dir[1] == 0:
                return [dir]
            return [np.array([1, 0]), np.array([-1, 0])]
```

```python
def check(val, shape):
    return (val >= 0).all() and (val < np.array(shape)).all()
```

```python
def dir_dict(dir):
    match tuple(dir):
        case (1, 0):
            return 0
        case (0, 1):
            return 1
        case (-1, 0):
            return 2
        case (0, -1):
            return 3
```

```python
from collections import deque

def compute(map, start):
    q = deque()
    q.append(start)
    
    visited = np.zeros((*map.shape, 4), dtype=bool)
    visited[tuple([*start[0], dir_dict(start[1])])] = True
    
    while len(q):
        current = q.popleft()
        dirs = nx(map[tuple(current[0])], current[1])
        for dir in dirs:
            new_current = np.array([current[0] + dir, dir])
            if check(new_current[0], map.shape) and not visited[tuple([*new_current[0], dir_dict(new_current[1])])]:
                visited[tuple([*new_current[0], dir_dict(new_current[1])])] = True
                q.append(new_current)
    
    return visited.any(axis=2).astype(int).sum()

compute(map, np.array([[0, 0], [0, 1]]))
```
