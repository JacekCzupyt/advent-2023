```python
seeds = [int(x) for x in input.split("\n\n")[0].split(": ")[1].split(" ")]
seeds
```

```python
map_strings = input.split("\n\n")[1:]
maps = [[[int(z) for z in y.split(" ")] for y in x.split(":\n")[1].split("\n")] for x in map_strings]
maps
```

```python
def map_seed(seed, maps):    
    for map in maps:
        for range in map:
            sub = seed - range[1]
            if sub < range[2] and sub >= 0:
                seed = range[0] + sub
                break
    
    return seed
```

```python
min = map_seed(seeds[0], maps)

for seed in seeds:
    res = map_seed(seed, maps)
    if res < min:
        min = res
        
min
```

```python
seed_rng = np.reshape(seeds, (-1, 2))
seed_rng
```

```python
from dataclasses import dataclass


@dataclass
class MapRange:
    start: int
    end: int
    offset: int
```

```python
new_maps = [
    [
        MapRange(start=range[1], end=range[1]+range[2], offset=range[0]-range[1])
        for range in map
    ]
    for map in maps
]
new_maps
```

```python
def single_step_ranges(ranges, map):
    results = []
    for range in ranges:
        current = range[0]
        remaining = range[1]
        i = 0
        
        while i<len(map) and map[i].end < current:
            i+=1
            
        while remaining > 0:
            if i>=len(map):
                results.append([current, remaining])
                remaining = 0
            elif map[i].start > current:
                results.append([current, np.min([remaining, map[i].start - current])])
                remaining -= map[i].start - current
                current = map[i].start
            else:
                results.append([current + map[i].offset, np.min([remaining, map[i].end - current])])
                remaining -= map[i].end - current
                current = map[i].end
                i+=1
                
    return results
            
                
```

```python
ranges = seed_rng
for map in new_maps:
    ranges = single_step_ranges(ranges, map)
    
ranges
```
