```python
data = input.split("\n")
data = ["." + d + "." for d in data]
data = ["." * len(data[0])] + data + ["." * len(data[0])]
data
```

```python
res = []
bad_res = []

for l, line in enumerate(data):
    i = 0
    k = 0
    while i < len(line):
        if line[i].isnumeric():
            k+=1
        elif k > 0:
            value = int(line[i-k:i])
            radius = data[l-1][i-k-1:i+1] + data[l][i-k-1:i+1] + data[l+1][i-k-1:i+1]
            if len(re.findall(r"[^0-9.]", radius)) > 0:
                res.append(value)
            else:
                bad_res.append(value)
            k = 0
        i+=1
res, bad_res
```

```python
from collections import defaultdict

res2 = defaultdict(lambda: [])

for l, line in enumerate(data):
    i = 0
    k = 0
    while i < len(line):
        if line[i].isnumeric():
            k+=1
        elif k > 0:
            value = int(line[i-k:i])
            radius = data[l-1][i-k-1:i+1] + data[l][i-k-1:i+1] + data[l+1][i-k-1:i+1]
            for f in re.finditer(r"\*", radius):
                pos = f.start()
                loc = (pos//(len(radius)//3) + l - 1, pos%(len(radius)//3) + i - k - 1)
                res2[loc].append(value)
            else:
                bad_res.append(value)
            k = 0
        i+=1

print(res2)
```
