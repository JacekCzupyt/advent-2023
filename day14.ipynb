```python
def calc(map):
    res = 0
    
    n = map.shape[0]
    for i in range(map.shape[1]):
        k = n
        for j in range(n):
            if map[j, i] == 'O':
                res += k
                k -= 1
            elif map[j, i] == '#':
                k = n - j - 1
    
    return res

calc(map)
```

## Part 2

```python
def roll(map):
    n = map.shape[0]
    for i in range(map.shape[1]):
        k = 0
        for j in range(n):
            if map[j, i] == 'O':
                map[j, i], map[k, i] = map[k, i], map[j, i]
                k += 1
            elif map[j, i] == '#':
                k = j + 1
    
    return map

roll(map)
```

```python
def map_to_str(map):
    return "\n".join(["".join(x) for x in map])

print(map_to_str(map))
```

```python
dict = {}
m = 1000000000
i = 0
while i < m:
    if (s := map_to_str(map)) in dict:
        print(dict[s], i)
        loop = i-dict[s]
        i += loop * ((m - i) // loop)
        print(i)
    dict[s] = i
    for _ in range(4):
        map = roll(map)
        map = np.rot90(map, -1)
    i+=1
```
