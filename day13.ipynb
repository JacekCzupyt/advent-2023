```python
def find_reflection(map):
    n = len(map)
    k = 1
    l = 0
    p = np.zeros(n, dtype=int)
    res = -1

    for i in range(n - 1):
        if i >= l:
            k, l = i + 1, i
            while k > 0 and l < n - 1 and map[k - 1] == map[l + 1]:
                l += 1
                k -= 1
            p[i] = (l - k + 1) // 2
        else:
            p[i] = max(0, min(l - i, p[k + i - l]))
            while p[i] < i and i + p[i] < n - 1 and map[i - p[i]] == map[i + p[i] + 1]:
                p[i] += 1
            if i + p[i] > l:
                k, l = i - p[i] + 1, i + p[i]

        if (k == 0 or l == n - 1) and res == -1:
            res = i

    return p, res


find_reflection(maps[0])
```

```python
res = []

for map in maps:
    res1 = find_reflection(map)
    res2 = find_reflection(["".join([map[i][j] for i in range(len(map))]) for j in range(len(map[0]))])
    res.append((res1[1] + 1, res2[1] + 1))

res = np.array(res)
res
```

```python
def find_smudged_reflection(map):
    n = len(map)
    p = np.zeros(n-1, dtype=int)
    res = -1

    for i in range(n - 1):
        diff = 0
        while p[i] < i+1 and p[i] < n-i-1 and diff + (d := (map[i-p[i]] != map[i+p[i]+1]).sum()) <= 1:
            p[i] += 1
            diff += d
        if diff == 0:
            p[i] = 0
        
            
    res = np.where(np.logical_or(p == np.arange(n-1) + 1, p == np.arange(n-1, 0, -1)))[0]
    res = res[0] if len(res) > 0 else -1
        
    return p, res + 1


find_smudged_reflection(maps[0])
```

```python
res = []

for map in maps:
    res1 = find_smudged_reflection(map)
    res2 = find_smudged_reflection(map.T)
    res.append((res1[1], res2[1]))

res = np.array(res)
res
```
