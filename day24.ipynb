```python
def check_intersect(h1, h2, num_axis=3, spec=False):
    h = np.array([h1, h2])
    vel_mat = np.array([[h[0, 1, ax], -h[1, 1, ax]] for ax in range(num_axis)], dtype=np.int64)
    pos_vec = np.array([[h[1, 0, ax] - h[0, 0, ax]] for ax in range(num_axis)], dtype=np.int64)
    try:
        t = scipy.linalg.solve(vel_mat, pos_vec)
        if spec and (t < 0).any():
            return False, None, None
        pos = (h[0, 0] + t[0] * h[0, 1])[:2]
        if spec and ((pos < test_area[:, 0]).any() or (pos > test_area[:, 1]).any()):
            return False, None, None
        return True, t, pos
    except:
        return False, None, None
```

```python
total = 0

for i in range(len(hail)):
    for j in range(i+1, len(hail)):
        res, t, pos = check_intersect(hail[i], hail[j], 2, True)
        if res:
            total+=1
            print(f"Hail: {i}-{j}, T: {t[:, 0]}, pos: {pos}")
            
total
```
