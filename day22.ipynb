```python
map_shape = (max([end[0] for block in blocks for end in block])+1, max([end[0] for block in blocks for end in block])+1)
height_map = np.zeros(map_shape, dtype=int)
block_map = np.ones_like(height_map, dtype=int) * -1
```

```python
is_supported_by = [set() for _ in blocks]
supports = [set() for _ in blocks]


for k, block in enumerate(blocks):
    r = [(block[:, 0].min(), block[:, 0].max()+1), (block[:, 1].min(), block[:, 1].max()+1)]
    support_level = np.max(height_map[r[0][0]:r[0][1], r[1][0]:r[1][1]], axis=None)
    
    if support_level > 0:
        for i in range(r[0][0], r[0][1]):
            for j in range(r[1][0], r[1][1]):
                if height_map[i, j] == support_level:
                    is_supported_by[k].add(block_map[i, j])
                    supports[block_map[i, j]].add(k)
    
    block[:, 2] += support_level + 1 - block[:, 2].min()
    height_map[r[0][0]:r[0][1], r[1][0]:r[1][1]] = block[:, 2].max()
    block_map[r[0][0]:r[0][1], r[1][0]:r[1][1]] = k
    
block_map
```

```python
total = 0

for i in range(len(supports)):
    if len(supports[i]) == 0:
        total += 1
    elif all([len(is_supported_by[j]) != 1 for j in supports[i]]):
        total += 1
        
total
```
