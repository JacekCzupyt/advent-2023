```python
lines = input.split("\n")
start = (0, 0)

map = np.zeros((len(lines) + 2, len(lines[0]) + 2))
for j, line in enumerate(lines):
    for i, c in enumerate(line):
        if c in ['.', 'S']:
            map[j + 1, i + 1] = 1
        if c == 'S':
            start = (j + 1, i + 1)
```

```python
def draw_dist(map, start):
    dist = np.ones_like(map) * 999

    q = deque()
    q.append(start)
    dist[start] = 0
    while len(q):
        node = q.popleft()
        for dir in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
            new_node = tuple(np.array(node) + np.array(dir))
            if map[new_node] == 1 and dist[new_node] == 999:
                q.append(new_node)
                dist[new_node] = dist[node] + 1

    return dist
```

```python
dist = draw_dist(map, start)
dist
```

```python
# manage_full

big_map = np.append(map[1:-1, 1:-1], map[1:-1, 1:-1], axis=1)
big_map = np.append(big_map, big_map, axis=0)
big_map = np.pad(big_map, 1)
big_map = big_map.astype(int)
big_map.shape
```

```python
def get_diag(big_map, num_steps, start, startval):
    res1 = draw_dist(big_map, start)
    # res1m = np.zeros_like(res1) + res1 * np.logical_and(res1 != 999, (res1 % 2) == (num_steps % 2))
    res1m = (((num_steps - startval - res1) // (true_shape[0] * 2)) + 1) * np.logical_and(res1 != 999, ((res1 + startval) % 2) == (num_steps % 2))
    res1m = np.max([res1m, np.zeros_like(res1m)], axis=0)
    res1m[res1m==999] = 0
    res1m = (res1m + 1) * res1m // 2
    return res1m.astype(np.int64).sum()


get_diag(big_map, num_steps, (1, 1), dist[-2, -2]+2)
```

```python
diags = sum([
    get_diag(big_map, num_steps, (1, 1), dist[-2, -2]+2),
    get_diag(big_map, num_steps, (1, -2), dist[-2, 1]+2),
    get_diag(big_map, num_steps, (-2, -2), dist[1, 1]+2),
    get_diag(big_map, num_steps, (-2, 1), dist[1, -2]+2),
])
diags
```

```python
def get_straight(map, num_steps, starts):
    offset = min([start["dist"] for start in starts])
    dists = [
        draw_dist(map, start["pos"]) + (map == 1) * (start['dist'] - offset)
        for start in starts
    ]
    res1 = np.min(dists, axis=0)
    res1m = res1 + true_shape[0] * np.logical_and(res1 != 999, ((res1 + offset) % 2) != (num_steps % 2))
    res1m = (((num_steps - offset - res1m) // (true_shape[0] * 2)) + 1) * (res1m != 999)
    res1m = np.max([res1m, np.zeros_like(res1m)], axis=0)
    # res1m[res1m == 999] = 0
    return res1m.astype(np.int64).sum()

get_straight(map, num_steps, [{'pos': (1, 1), 'dist': dist[1, -2]+1}, {'pos': (-2, 1), 'dist': dist[-2, -2]+1}])
```

```python
straights = sum([
    get_straight(map, num_steps, [{'pos': (66, 1), 'dist': dist[66, -2]+1}]),
    get_straight(map, num_steps, [{'pos': (66, -2), 'dist': dist[66, 1]+1}]),
    get_straight(map, num_steps, [{'pos': (1, 66), 'dist': dist[-2, 66]+1}]),
    get_straight(map, num_steps, [{'pos': (-2, 66), 'dist': dist[1, 66]+1}]),
])
straights
```
