```python
order = ["J"] + [str(x) for x in range(2, 10)] + ["T", "Q", "K", "A"]
order_dict = {v: i for i, v in enumerate(order)}
order_dict
```

```python
def get_hand_type(hand):
    simple_hand = [char for char in list(hand[0]) if char != "J"]
    simple_type = sorted(np.unique(simple_hand, return_counts=True)[1], reverse=True) or [0]
    simple_type[0] += 5 - len(simple_hand)
    return tuple(simple_type)

hand_values = [(
    get_hand_type(hand),
    tuple([order_dict[card] for card in list(hand[0])])
) for hand in cards]
hand_values
```

```python
hand_ranking1 = [i[0] for i in sorted(enumerate(hand_values), key=lambda x: x[1])]
hand_ranking = [i[0] for i in sorted(enumerate(hand_ranking1), key=lambda x: x[1])]
np.array(hand_ranking) + 1
((np.array(hand_ranking) + 1) * np.array([int(hand[1]) for hand in cards])).sum()
```
